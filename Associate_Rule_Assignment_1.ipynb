```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Import Dataset
book=pd.read_csv('/content/book.csv')
book
```

```python
# Data preprocessing not required as it is already in transaction format
```

```python
'''
Apriori Algorithm
1. Association rules with 10% Support and 70% confidence
'''
```

```python
# With 10% Support
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# Most frequent itemsets based on support 
frequent_itemsets.sort_values('support', ascending = False, inplace = True)
frequent_itemsets
```

```python
# with 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
'''
 A leverage value of 0 indicates independence. Range will be [-1 1]
 A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

'''
```

```python
rules.sort_values('lift', ascending = False).head(10)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
rules.plot(x = "support", y = "confidence", c = rules.lift, kind="scatter", s = 12, cmap = plt.cm.coolwarm)
```

```python
# With 5% Support
frequent_itmsets=apriori(book,min_support=0.05,use_colnames=True)
frequent_itmsets
```

```python
# With 80% confidence
ruless=association_rules(frequent_itmsets,metric='lift',min_threshold=0.8)
ruless
```

```python
ruless[ruless.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(ruless['support'],ruless['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
